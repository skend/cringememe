```python
import requests
import json
import time
import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime
from scipy.ndimage import gaussian_filter1d
from dateutil import parser
```

```python
import emoji

def count_emoji(s):
    return int(len([c for c in emoji.demojize(s) if c == ':']) / 2)
```

```python
url = "https://www.reddit.com/user/propertyoflenore/comments/.json?limit=100"
```

```python
comments = []
def get_more_comments(after):
    res = requests.get(url + ';after=' + after, headers = {'User-Agent': 'NikosamHater69'})
    try:
        jsonres = json.loads(res.text)
        for child in jsonres['data']['children']:
            if child['data']['subreddit'] == 'Destiny':
                comments.append((datetime.datetime.fromtimestamp(child['data']['created']), count_emoji(child['data']['body'])))
        return jsonres['data']['after'], 1
    except Exception as e:
        print(e)
        return after, 0
```

```python
after = ''
count = 0
while count < 4:
    after, success = get_more_comments(after)
    count += success
    time.sleep(2)
```

```python
d = {}
for comment in comments:
    trunc_date = str(comment[0].replace(hour=0, minute=0, second=0, microsecond=0))
    if trunc_date in d.keys():
        d[trunc_date] += comment[1]
    else:
        d[trunc_date] = 0
```

```python
x = []
y = []
for k in d.keys():
    x.append(parser.parse(k))
    y.append(d[k])
```

```python
ysmoothed = gaussian_filter1d(y, sigma=2)
```

```python
plt.style.use('seaborn-poster')
plt.plot(x, ysmoothed)
plt.xlabel('date', fontsize=15)
plt.ylabel('emojis per day', fontsize=15)
plt.rcParams['figure.figsize'] = [10/2.54, 8/2.54]
dtFmt = mdates.DateFormatter('%b-%d')
plt.gca().xaxis.set_major_formatter(dtFmt)
plt.title("nikosam's emoji usage", fontsize=15)
plt.axvline(x=datetime.datetime.fromtimestamp(1643140077), color='red')
plt.show()
```
